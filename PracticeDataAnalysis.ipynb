```python
# Q33: Top 10 highest value orders
# df.nlargest(10,'total_amount').sort_values('total_amount')
df.sort_values('total_amount', ascending=False).head(10)
```

```python
# Q34: Bottom 10 lowest value orders
df.sort_values('total_amount').head(10)
```

```python
# Q35: Sort by multiple columns
df.sort_values(['total_amount','discount_percent'], ascending=[True,False])
```

```python
# Q36: Total sales by category
df.groupby('category').sum()['total_amount']
```

```python
# Q37: Average order value by region
df.groupby('region')['total_amount'].mean()
```

```python
# Q38: Count of orders by customer segment
df.groupby('customer_segment').count()['order_id']
```

```python
# Q39: Total profit by product category
df.groupby('category').sum()['profit']
```

```python
# Q40: Top 5 customers by total spending
df.groupby('customer_name').sum()['total_amount'].sort_values(ascending=False).head(5)
```

```python
# Q41: Average discount by category
df.groupby('category')['discount_percent'].mean()
```

```python
# Q42: Sales by payment method
df.groupby('payment_method').sum()['total_amount']
```

```python
# Q43: Profitable vs unprofitable products
totalprofit = df.groupby('product_name').sum()['profit'].sort_values(ascending=False)

profit_prod = totalprofit[totalprofit > 0]
loss_prod = totalprofit[totalprofit < 0]

print(profit_prod)
print()
print(loss_prod)
```

```python
# Q44: Region-wise category performance
df.groupby(['region','category']).sum()['total_amount']
```

```python
# Q45: Monthly sales trend
df['order_month'] = pd.to_datetime(df['order_date']).dt.to_period('M')
df.groupby('order_month').sum()['total_amount']
```

```python
# Q46: Products with high quantity but low value
df.loc[ (df['quantity'] > 10) & (df['total_amount'] < 100) ]
```

```python
# Q47: Best performing sales channel by region
df.groupby(['sales_channel','region'])['total_amount'].sum()
```

```python
# Q48: Corporate customers high value orders
df.loc[ (df['customer_segment']=='Corporate') & (df['total_amount'] > 500) ]
```

```python
# Q49: Profit margin analysis
#Profit margin = (Profit / Total Amount) Ã— 100%
df['profit_margin'] = (df['profit']/df['total_amount']) * 100
print(df.groupby('category')['profit_margin'].mean())
print()
print(df.groupby('region')['profit_margin'].mean())
```

```python
# Q50: Customer frequency analysis
df.groupby('customer_name').count()['order_id'].rename('total_order_count').sort_values(ascending=False)
```
