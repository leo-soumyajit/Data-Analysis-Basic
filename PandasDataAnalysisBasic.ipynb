Pandas Data Analysis Basic

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data.csv')
print("CSV Loaded into DataFrame 'df' . First 5 rows:")
print(df.head()) #first 5 data values
# df.head()
```

```python
print(df.tail()) #last 5 data values
```

```python
print("Shape", df.shape)
print("Size",df.size)
print("Dimension",df.ndim)
print("Column Names",df.columns.tolist())
print("Index",df.index.tolist())
print("Data Types \n",df.dtypes)
print("\nInfo")
df.info()
```

```python
df.describe()
```

Select and filter data

```python
print("\n --- 1 Selecting Name and Department columns ---")
selected_column = df[["name", "department"]]
print(selected_column)
```

```python
print("\n --- 2 Selecting row with index 5 using .loc ---")
print(df.loc[5]) #index 5 all data
```

```python
print("\n --- 3 Selecting first 3 rows and columns from index 0 to 4 using .iloc ---")
print(df.iloc[0:3,0:5]) #rows 0-2 and col 0-4
```

```python
print("\n --- 4 Filtering rows where salary > 70000 using boolean indexing ---")
print(df [ df['salary']> 70000 ] )
```

```python
print("\n --- 5 Filtering rows where department is Engineering and Project Completed > 10 ---")
print(df [ (df['department'] == 'Engineering') & (df['projects_completed'] > 10) ] )
```

```python
print("\n --- 6 Filtering rows where city is Neywork ---")
print(df [ df['city'] == 'New York' ])
```

```python
print("\n --- 7 Filtering for remote employees ---")
print( df[ df['is_remote'] == True] )
```

```python
print("\n --- 8 Filtering for employees with age bw 30 and 35 ---")
print( df[ (df['age']>=30) & (df['age'] <= 35) ] )
```

```python
print("\n --- 9 Selecting name and salary of employee in sales---")
print( df.loc[df['department'] == 'Sales', ['name', 'salary']] )
```

```python
print("\n --- 10 Filtering rows where department is either Marketing or HR---")
print( df[df['department'].isin(['Marketing','HR'])] )
```

```python
print("\n --- 11 Filtering employees whose name contains John---")
print( df[df['name'].str.contains('John', case=False, na = False)] )
```

```python
print("\n --- 12 Selecting name department and salary for employees with performance_score> 9 and projects_completed >= 15 ---")
print( df.loc[ (df['performance_score'] > 9) & (df['projects_completed'] >= 15) , ['name','department','salary'] ] )
```

```python
print("\n --- 13 selecting all employees except those in Sales department---")
print( df.loc[ df['department'] != 'Sales' ] )
```

**Missing Value Basic**

```python
print("---1. Indentifing missing values in df----")
print(df.isnull().sum())
```

```python
print("---2 Filling missing salary and perfomance_score with mean and city with mode---")
df['salary'] = df['salary'].fillna(df['salary'].mean())
df['performance_score'] = df['performance_score'].fillna(df['performance_score'].mean())


#mode for city
df['city'] = df['city'].fillna(df['city'].mode()[0])

print("Missing values after filling")
print(df.isnull().sum())

```

```python
print(df)
```

**Group and Agreegation**

```python
print("--- 1. Mean of numarical columns grouped by department---")
print(df.groupby('department').mean(numeric_only=True))
```

```python
print("--- 2. Sum of numarical columns grouped by department---")
print(df.groupby('department').sum(numeric_only=True))
```

```python
print("--- 3. Count of numarical columns grouped by department---")
print(df.groupby('department').size().reset_index(name="Emloyee Count"))
```

**Visualisation**

```python
import matplotlib.pyplot as plt
```

```python
print("--- 1 Calculate avg sal per department")
avg_sal_per_department = df.groupby('department')['salary'].mean()
print(avg_sal_per_department)
```

```python
print("--- 2 Create a bar chart of avg sal per department")
plt.figure(figsize=(10,6))
avg_sal_per_department.plot.bar(color="skyblue")


#add title and lables
plt.title('Average Salary per Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45,ha='right') #Rotate x-axis lables for better readability
plt.tight_layout()
plt.show()
```
